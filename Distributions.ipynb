# Examples of Data Distributions

## Uniform Distribution

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


# Declaramos un arreglo de distribucion uniforme, con valores que oscilan entre -10 y 10 y 100000 muestras

values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
```

## Normal / Gaussian

Visualice la función de densidad de probabilidad:

```python
from scipy.stats import norm
import matplotlib.pyplot as plt


# creamos un arreglo de valores entre -3 y 3, que se van incrementando en 0.001
x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))  # norm.pdf es la funcion de densidad
```

Genera algunos números aleatorios con una distribución normal. "mu" es la media deseada, "sigma" es la desviación estándar:

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 5.0
sigma = 2.0

# creamos un arreglo con inicio en 5.0, una desv-std de 2.0 con 10000 muestras
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

## Exponential PDF (funcion distribucion y probabilidad exponencial) / "Power Law"

*Se utiliza principalmente para "mostrar" una caida exponencial en analisis que se esta realizando*

```python
from scipy.stats import expon
import matplotlib.pyplot as plt


# Creamos un arreglo con valores que flutuan entre 0 y 10 se mueven cada 0.001

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

## Binomial Probability Mass Function 
**Funcion de masa de probabilidad binominal**

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.5 # parametros

# Creamos un arreglo con valores que flutuan entre 0 y 10 se mueven cada 0.001

x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

## Poisson Probability Mass Function
**Funcion de masa de probabilidad poison**

Ejemplo: Mi sitio web recibe un promedio de 500 visitas por día. ¿Cuáles son las probabilidades de obtener 550?

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 500 # media visitantes

# Creamos un arreglo con valores que flutuan entre 400 y 600 se mueven cada 0.5
x = np.arange(400, 600, 0.5)

plt.plot(x, poisson.pmf(x, mu))

print('Segun muestra la grafica la posibilidad de recibir 550 vistantes en un dia es del 0.002 aprox')
```
