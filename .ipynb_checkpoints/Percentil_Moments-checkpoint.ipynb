# Percentiles
*En un conjunto de datos, ¿cuál es el punto en el que el X% de los valores es menor que ese valor?*

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


# generamos un arreglo centrados en cero con una media de cero, una desv std de 0.5 y 10000 muestras
vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

```python
print('El valor del percentil 50 es..',np.percentile(vals, 50))
print('El valor del percentil 90 es..',np.percentile(vals, 90))
print('El valor del percentil 20 es..',np.percentile(vals, 20))
```

# Momentos: Media, Varianza, Sesgo, Curtosis

*Medidas cuantitativas de la forma de una función de densidad de probabilidad. Matemáticamente son un poco difíciles de entender. Pero intuitivamente, es mucho más simple en estadísticas. El primer momento es la media, el segundo momento es la varianza*

**El tercer momento es "sesgado"**

- ¿Qué tan "desequilibrada" es la distribución?
- Una distribución con una cola más larga a la izquierda estará sesgada hacia la izquierda y tendrá un sesgo negativo

**El cuarto momento se llama "curtosis"**

- ¿Qué tan gruesa es la cola y qué tan afilada es el pico, en comparación con una distribución normal?


*Cree un conjunto de datos aleatorios distribuidos aproximadamente normalmente:*

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# generamos un arreglo centrados en cero con una media de cero, una desv std de 0.5 y 10000 muestras
vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

*El primer momento es el medio; Estos datos deben promediar alrededor de 0:*

```python
print('El primer momento corresponde a ..',np.mean(vals))
print('El segundo momento corresponde a ..',np.var(vals))
```

**El tercer momento está sesgado: dado que nuestros datos están bien centrados alrededor de 0, debería ser casi 0:**

```python
import scipy.stats as sp
#sp.skew(vals)
print('El tercer momento corresponde a ..',sp.skew(vals))
```

**El cuarto momento es la "curtosis", que describe la forma de la cola. Para una distribución normal, esto es 0:**

```python
print('El cuarto momento corresponde a ..',sp.kurtosis(vals))
```
