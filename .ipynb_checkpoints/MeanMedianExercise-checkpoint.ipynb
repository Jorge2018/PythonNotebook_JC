# Exercise: Mean & Median Customer Spend

Aquí hay un código que generará algunos datos aleatorios de comercio electrónico; solo una matriz de la cantidad total gastada por transacción. Seleccione el bloque de código y presione "reproducir" para ejecutarlo:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 20.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

Ahora, encuentre la media y la mediana de estos datos. En el bloque de código siguiente, escriba su código y vea si su resultado tiene sentido:

```python
print('La media de ingresos es..',np.mean(incomes))
print('La mediana de ingresos  es..',np.median(incomes))
```
