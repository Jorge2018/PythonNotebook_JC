# Standard Deviation and Variance

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# El arreglo ingresos lo centramos en 100, con una desv estandar de 20 y 10000 apariciones (ptos de datos)

incomes = np.random.normal(100.0, 20.0, 10000)

plt.hist(incomes, 50)
plt.show()
```

```python
print('La variacion estandar del arreglo de datos es..',incomes.std())
```

```python
print('La varianza del arreglo de datos es..',incomes.var())
```
